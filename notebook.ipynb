```python

import pandas as pd
import tabula

# Read PDF tables
tables = tabula.read_pdf(r"C:\Users\praga\Downloads\pre-monsoon_1994-2003.pdf", pages="all", multiple_tables=True)

# Save each table to Excel
with pd.ExcelWriter("output.xlsx") as writer:
    for i, table in enumerate(tables):
        table.to_excel(writer, sheet_name=f"Table_{i+1}", index=False)
```
